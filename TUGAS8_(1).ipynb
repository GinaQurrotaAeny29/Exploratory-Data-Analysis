```python
# ========================================
# STEP 2: Import Library
# ========================================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ========================================
# STEP 3: Load Dataset
# ========================================
df = pd.read_csv('marvel_characters_dataset.csv')
```

```python
# ========================================
# STEP 4: Tampilkan 20 Data Teratas & Terbawah
# ========================================
print("=== 20 Data Teratas ===")
display(df.head(20))

print("\n=== 20 Data Terbawah ===")
display(df.tail(20))
```

```python
# ========================================
# STEP 5: Tampilkan Tipe Data per Kolom
# ========================================
print("\n=== Tipe Data Setiap Kolom ===")
print(df.dtypes)
```

```python
# ========================================
# STEP 6: Statistik Deskriptif
# ========================================
print("\n=== Statistik Deskriptif ===")
display(df.describe(include='all'))
```

```python
# ========================================
# STEP 7: Cek Missing Value & Distribusi Data
# ========================================
print("\n=== Jumlah Missing Value Tiap Kolom ===")
print(df.isnull().sum())
```

```python
# Distribusi data numerik (contoh dengan histogram)
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribusi Data Kolom: {col}')
    plt.xlabel(col)
    plt.ylabel('Frekuensi')
    plt.tight_layout()
    plt.show()
```

```python
print("\n=== Korelasi Antara Kolom Numerik ===")
if len(numeric_cols) >= 3:
    corr_matrix = df[numeric_cols].corr()
    display(corr_matrix)

    # Visualisasi heatmap korelasi
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Heatmap Korelasi Kolom Numerik")
    plt.show()
else:
    print("Jumlah kolom numerik kurang dari 3 untuk analisis korelasi.")
```

```python
# ========================================
# STEP 9: Visualisasi Tambahan (Disesuaikan)
# ========================================

# Countplot berdasarkan Role
plt.figure(figsize=(10, 4))
sns.countplot(y='Affiliation', data=df, order=df['Affiliation'].value_counts().index)
plt.title('Distribusi Karakter Berdasarkan Affiliation')
plt.xlabel('Jumlah')
plt.ylabel('Affiliation')
plt.tight_layout()
plt.show()


# Boxplot pakai Power Level numerik
plt.figure(figsize=(12, 6))
sns.boxplot(x='Affiliation', y='Power Level', hue='Role', data=df)
plt.title('Distribusi Power Level Berdasarkan Affiliation dan Role')
plt.xlabel('Affiliation')
plt.ylabel('Power Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```
